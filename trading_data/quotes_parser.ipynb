```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.express as px

from moexalgo import Ticker
```

```python
START_DATE = "2023-01-01"
END_DATE = "2025-02-28"
```

#### SBER

```python
sber = Ticker("SBER")
sber_data = sber.candles(
    start=START_DATE,
    end=END_DATE,
    period="1d",
)
```

```python
sber_data
```

```python
# fig = px.line(
#     sber_data, x=sber_data.begin, y="close", title="Цена закрытия акций Сбербанка"
# )
# fig.update_layout(
#     xaxis_title="Дата",
#     yaxis_title="Цена, руб.",
#     template="plotly_white",
#     hovermode="x unified",
# )
# fig.show()

fig = px.line(
    sber_data, 
    x=sber_data.begin, 
    y="close", 
    title="Цена закрытия акций SBER",
    color_discrete_sequence=['green']  # Изменение цвета линии на зеленый
)

fig.update_layout(
    xaxis_title="Дата",
    yaxis_title="Цена, руб.",
    template="plotly_white",
    hovermode="x unified",
    width=700,  # Уменьшаем ширину графика
    height=400,  # Уменьшаем высоту графика
    margin=dict(l=50, r=50, t=50, b=50)  # Уменьшаем отступы
)

fig.show()
```

```python
sber_data.to_csv('data/sber_data.csv', index=False)
```
